```python
class DataIter():
    def __init__(self, df):
        self.df = df
        self.size = len(self.df)
        self.epochs = 0
        self.shuffle()
    def shuffle(self):
        self.df = self.df.sample(frac=1).reset_index(drop=True)
        self.cursor = 0
    def next_batch(self, n):
        if self.cursor + n > self.size:
            #res = self.df.iloc[self.cursor:self.size]
            self.epochs += 1
            self.shuffle()

        res = self.df.iloc[self.cursor:(self.cursor+n)]
        self.cursor += n
        maxlen = max(res['length'])
        #print(maxlen)
        #print(res.shape)
        x = np.zeros([n, maxlen], dtype=np.int32)
        for i, x_i in enumerate(x):
            x_i[:res['length'].values[i]] = res['lab'].values[i].copy()
            #res['lab'].values[i] = x_i 
        return x, res['y'], res['length'],res.index
#train_data = DataIter(trainset)
#train_data.next_batch(5)
test_data = DataIter(testset)
test_data.next_batch(5)
```

```python
def reset_graph():
    if 'sess' in globals() and sess:
        sess.close()
    tf.reset_default_graph()

#hyperparam
LR = 0.001 # learning rate
ITERATION = 10000
BATCH_SIZE = 300
EMBED_SIZE = 100
CELL_SIZE = 10
KEEP_PROB = 0.7
MAX_LENGTH = vocab_size# maximum vocab
TIME_STEP = sen_size # maximun sentence length

# input
def build_graph(
    vocab_size = MAX_LENGTH+1,
    cell_size = CELL_SIZE,
    batch_size = BATCH_SIZE,
    embed_size = EMBED_SIZE,
    num_classes = 2,
    keep_prob = KEEP_PROB):
    reset_graph()
    
    # placeholders
    x = tf.placeholder(tf.int32, [batch_size, None]) # [batch_size, num_steps]
    seqlen = tf.placeholder(tf.int32, [batch_size])
    y = tf.placeholder(tf.int32, [batch_size])
    keep_prob = tf.constant(KEEP_PROB)
    
    # Embedding layer
    embeddings = tf.get_variable('embedding_matrix',[vocab_size, embed_size])
    rnn_inputs = tf.nn.embedding_lookup(embeddings, x)
    print(rnn_inputs.get_shape)
    
    # Input layer
    rnn_inputs = tf.layers.dense(rnn_inputs, units=cell_size)
    print(rnn_inputs.get_shape)
    
    # RNN
    cell = tf.nn.rnn_cell.GRUCell(cell_size)
    init_state = tf.get_variable('init_state',[1,cell_size],
                                initializer = tf.constant_initializer(0.0))
    init_state = tf.tile(init_state, [batch_size, 1])
    rnn_outputs , final_state = tf.nn.dynamic_rnn(cell, rnn_inputs, sequence_length = seqlen, initial_state = init_state)
    
    # Add dropout
    rnn_outputs = tf.nn.dropout(rnn_outputs, keep_prob)
    #print(rnn_outputs.get_shape)
    
    # Select last output
    last_rnn_output = tf.gather_nd(rnn_outputs, tf.stack([tf.range(batch_size), seqlen-1], axis=1))
    last_rnn_output = tf.layers.dense(last_rnn_output, units=30)
    #print(last_rnn_output.get_shape)
    
    
    
    # Cost
    with tf.variable_scope('softmax'):
        W = tf.get_variable('W', [30, num_classes])
        b = tf.get_variable('b', [num_classes], initializer = tf.constant_initializer(0.0))
    logits = tf.matmul(last_rnn_output, W)+b
    print('logits:',logits.get_shape)
    preds = tf.nn.softmax(logits)
    #print(logits)
    correct = tf.equal(tf.cast(tf.argmax(preds, 1), tf.int32), y)
    accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
    print('y:',y)
    loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(labels = y, logits = preds))
    train_step = tf.train.AdamOptimizer(1e-4).minimize(loss)
    
    
    
    return{
        'x': x,
        'seqlen': seqlen,
        'y': y,
        'dropout': keep_prob,
        'loss': loss,
        'ts': train_step,
        'preds': preds,
        'accuracy': accuracy
    }

build_graph()
```

```python
def train_graph(graph, train, test, batch_size = BATCH_SIZE, num_epochs=50, iterator = DataIter):
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        tr = iterator(train)
        step, accuracy = 0,0
        tr_losses = []
        current_epoch = 0
        while current_epoch < num_epochs:
            step += 1
            batch = tr.next_batch(batch_size)
            feed = {graph['x']: batch[0], graph['y']: batch[1], graph['seqlen']:batch[2]}
            accuracy_,_ = sess.run([graph['accuracy'],graph['ts']], feed_dict = feed)
            accuracy += accuracy_
            
            if tr.epochs > current_epoch:
                current_epoch += 1
                tr_losses.append(accuracy/step)
                step, accuracy = 0,0 
                print("Accuracy after epoch", current_epoch, " - tr:", tr_losses[-1])#, "- te:", te_losses[-1])
        cursor = 0
        while cursor <= len(test):
            if cursor+batch_size <= len(test):
                te = test.iloc[cursor:cursor+batch_size]
            else:
                te = pd.concat([test.iloc[cursor:len(test)],test.iloc[0:batch_size-len(test.iloc[cursor:len(test)])]])
            maxlen = max(te['length'])
            x = np.zeros([batch_size, maxlen], dtype=np.int32)
            #print(x.shape)
            for i, x_i in enumerate(x):
                x_i[:te['length'].values[i]] = te['lab'].values[i].copy()
            results = sess.run(graph['preds'],feed_dict={graph['x']:x,graph['seqlen']:te.length})
            if cursor == 0:
                prediction = pd.DataFrame(data=results,columns=['HillaryClinton','realDonaldTrump'])
            else:
                prediction = pd.concat([prediction, pd.DataFrame(data=results,columns=['HillaryClinton','realDonaldTrump'])])              
            cursor += batch_size
            
            
    return tr_losses,prediction
        
```
